**Парадокс Монти Холла — описание задачи**

Участнику предлагается выбрать одну из трёх дверей, за одной из которых находится приз.
После выбора ведущий, зная расположение приза, открывает одну из двух оставшихся дверей с пустым исходом.
Затем участнику предлагается либо оставить первоначальный выбор, либо сменить дверь.
Требуется определить, какая стратегия — сохранение выбора или его смена — обеспечивает большую вероятность выигрыша.

```python
import random

def monty_hall_simulation(trials=100_000):
    stay_win = 0
    switch_win = 0

    for _ in range(trials):
        doors = [0, 0, 0]
        prize = random.randint(0, 2)
        doors[prize] = 1

        choice = random.randint(0, 2)

        # Ведущий открывает дверь с козой, не выбранную игроком
        available_doors = [
            i for i in range(3)
            if i != choice and doors[i] == 0
        ]
        opened = random.choice(available_doors)

        # Дверь для смены
        switch_choice = next(
            i for i in range(3)
            if i != choice and i != opened
        )

        if doors[choice] == 1:
            stay_win += 1
        if doors[switch_choice] == 1:
            switch_win += 1

    return stay_win / trials, switch_win / trials

stay, switch = monty_hall_simulation()

print(f"Вероятность выигрыша при НЕ смене двери: {stay:.3f}")
print(f"Вероятность выигрыша при смене двери:    {switch:.3f}")
```
